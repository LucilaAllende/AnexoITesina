Los datos se obtuvieron de los scrapers que se corrieron en la página del SIPAS. Se obtuvieron 5 dataset: humedad, lluvia, temperatura mínima, temperatura media, y temperatura máxima. En cada uno de ellos se eliminaron las sondas que no eran de Chubut (unas 7 creo) y a cada sonda se le asignó el departamento al que pertenecía manualmente.

# Armar dataset de clima por departamento

Importamos las librerías necesarias.

```python
import pandas as pd
```

Cargamos el dataset de humedad, lo agrupamos por dpto y año

```python
datos_humedad = pd.read_csv('datos_sipas/Agrupado_por_dptos/humedad_estaciones.csv', sep=",")
datos_humedad
```

```python
datos_humedad_agrupados = datos_humedad.groupby(by=['Departamento','Año']).mean().round(2)
datos_humedad_agrupados
```

Cargamos el dataset de lluvia, lo agrupamos por dpto y año

```python
datos_lluvia = pd.read_csv('datos_sipas/Agrupado_por_dptos/lluvia_estaciones.csv', sep=",")
datos_lluvia
```

```python
datos_lluvia_agrupados = datos_lluvia.groupby(by=['Departamento','Año']).sum().round(2)
datos_lluvia_agrupados
```

Cargamos el dataset de temperatura máxima, lo agrupamos por dpto y año

```python
datos_tem_max = pd.read_csv('datos_sipas/Agrupado_por_dptos/temperatura_maxima_estaciones.csv', sep=",")
datos_tem_max
```

```python
datos_tem_max_agrupados = datos_tem_max.groupby(by=['Departamento','Año']).mean().round(2)
datos_tem_max_agrupados
```

Cargamos el dataset de temperatura media, lo agrupamos por dpto y año

```python
datos_tem_med = pd.read_csv('datos_sipas/Agrupado_por_dptos/temperatura_media_estaciones.csv', sep=",")
datos_tem_med
```

```python
datos_tem_med_agrupados = datos_tem_med.groupby(by=['Departamento','Año']).mean().round(2)
datos_tem_med_agrupados
```

Cargamos el dataset de temperatura mínima, lo agrupamos por dpto y año

```python
datos_tem_min = pd.read_csv('datos_sipas/Agrupado_por_dptos/temperatura_minima_estaciones.csv', sep=",")
datos_tem_min
```

```python
datos_tem_min_agrupados = datos_tem_min.groupby(by=['Departamento','Año']).mean().round(2)
datos_tem_min_agrupados
```

Lo siguiente es juntar todos los anteriores dataframes para armar el dataset

```python
import functools

dfs = [datos_humedad_agrupados, 
       datos_lluvia_agrupados, 
       datos_tem_max_agrupados, 
       datos_tem_med_agrupados,
       datos_tem_min_agrupados]

df_final = functools.reduce(lambda left,right: pd.merge(left,right,left_index=True,right_index=True), dfs)
df_final
```

Reiniciamos los índices, renombramos las columnas y convertimos el año en tipo fecha para que nos sea más fácil unirlo con el dataset de producción.

```python
datos_clima = df_final.reset_index()
datos_clima
```

```python
datos_clima.columns = ['departamento', 'anio', 'PROMEDIO_HUMEDAD', 'TOTAL_LLUVIA',
       'PROMEDIO_T_MAX', 'PROMEDIO_T_MED', 'PROMEDIO_T_MIN']
datos_clima['anio'] = pd.to_datetime(datos_clima['anio'], format="%Y")
datos_clima
```

```python
datos_clima.to_csv('datos_producidos/clima_chubut.csv', index=False)
```
