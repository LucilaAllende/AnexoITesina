# Armar dataset clima + producción

Anteriormente se obtuvieron los datos de clima de cada departamento de chubut, de la existencias ovinas y de la produccion de lana. Ahora vamos a combinar todo para poder empezar a entrenar a los modelos.

```python
import pandas as pd
```

```python
produccion_chubut = pd.read_csv('datos_producidos/produccion_chubut_2000_2020.csv', sep=",")
produccion_chubut
```

```python
clima_chubut = pd.read_csv('datos_producidos/clima_chubut.csv', sep=",")
clima_chubut
```

```python
merged_inner = pd.merge(left=produccion_chubut,right=clima_chubut)
merged_inner
```

```python
merged_inner.to_csv('datos_producidos/merged_inner.csv', index=False)
```

```python
merged_left_join = pd.merge(left=produccion_chubut,right=clima_chubut,how='left')
merged_left_join
```

```python
merged_left_join.to_csv('datos_producidos/merged_left_join.csv', index=False)
```

```python
merged_full_join = pd.merge(left=produccion_chubut,right=clima_chubut,how='outer')
merged_full_join
```

```python
merged_full_join.to_csv('datos_producidos/merged_full_join.csv', index=False)
```
