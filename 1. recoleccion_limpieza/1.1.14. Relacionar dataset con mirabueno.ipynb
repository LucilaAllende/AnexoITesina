# Combinar dataset con mirabueno

```python
import pandas as pd
```

Este dataset representa los datos climaticos y de produccin de un campo de Chubut

```python
datos_mirabueno = pd.read_csv('datos/datos_produccion_clima_mirabueno.csv', sep=",")
datos_mirabueno
```

Se lo va combinar con cada uno de los datasets de clima y produccion de los departamentos de chubut

```python
datos_chubut_1 = pd.read_csv('datos_producidos/merged_inner.csv', sep=",")
datos_chubut_2 = pd.read_csv('datos_producidos/merged_left_join.csv', sep=",")
datos_chubut_3 = pd.read_csv('datos_producidos/merged_full_join.csv', sep=",")
datos_chubut_1
```

Para poder hacer la combinacion sera necesario a los datasets quitarles columnas, renombrarlas pra que la informacion del mirabueno coincida con la que pudimos conseguir para los departamentos. 

Cambios en datos de mirabueno:
1. Se agregaron manualmente la columna provincia, departamento y total de ovinos. Vamos a considerar al campo "El mirabueno" como un departamento para tener mas datos.
2. Se elimaron columnas ya que no tenemos esa información para los departamentos.
3. Se renombraron las columnas
4. Se transformo el campo fecha en un datetime

```python
datos_mirabueno = datos_mirabueno.drop(
    columns=['esquila', 'kilo_lana_p/animal', 'porcentaje_paricion', 'rad_solar', 'acumulado_anual', 'acumulado_verano', 'vel_viento'])
datos_mirabueno['fecha'] = pd.to_datetime(datos_mirabueno['fecha'])
datos_mirabueno.columns = ['fecha', 'provincia', 'departamento', 'carneros', 'ovejas', 
                           'borregos', 'capones', 'cordero/as', 'total_ovinos',
                           'kilos_lana', 'rinde', 'finura', 'temp_media', 'temp_min',
                           'tem_max', 'humedad', 'cant_lluvia']
datos_mirabueno
```

Cambios en datos de produccion:
1. Se cambio la fecha 01-01 por 31-12 manualmente. Ya que tiene mucho sentido, es una manera de decir estos son los datos que se juntaron en el año.
2. Se renombraron las columnas

```python
datos_chubut_1.columns = ['fecha', 'provincia', 'departamento', 'carneros', 'ovejas', 'borregos',
                          'capones', 'cordero/as', 'total_ovinos', 'kilos_lana', 'finura',
                          'rinde', 'humedad', 'cant_lluvia', 'tem_max','temp_media', 'temp_min']

datos_chubut_2.columns = ['fecha', 'provincia', 'departamento', 'carneros', 'ovejas', 'borregos',
                          'capones', 'cordero/as', 'total_ovinos', 'kilos_lana', 'finura',
                          'rinde', 'humedad', 'cant_lluvia', 'tem_max','temp_media', 'temp_min']

datos_chubut_3.columns = ['fecha', 'provincia', 'departamento', 'carneros', 'ovejas', 'borregos',
                          'capones', 'cordero/as', 'total_ovinos', 'kilos_lana', 'finura',
                          'rinde', 'humedad', 'cant_lluvia', 'tem_max','temp_media', 'temp_min']

datos_chubut_1
```

Ahora se va a concatenar los datos del mirabueno a los otros tres dataset

```python
datos_1 = pd.concat([datos_chubut_1, datos_mirabueno])
datos_1['fecha'] = pd.to_datetime(datos_1['fecha'])
datos_1 = datos_1.sort_values(['departamento', 'fecha'])
datos_1
```

```python
datos_1.to_csv('datos_producidos/datos_1.csv', index=False)
```

```python
datos_2 = pd.concat([datos_chubut_2, datos_mirabueno])
datos_2['fecha'] = pd.to_datetime(datos_2['fecha'])
datos_2 = datos_2.sort_values(['departamento', 'fecha'])
datos_2
```

```python
datos_2.to_csv('datos_producidos/datos_2.csv', index=False)
```

```python
datos_3 = pd.concat([datos_chubut_3, datos_mirabueno])
datos_3['fecha'] = pd.to_datetime(datos_3['fecha'])
datos_3 = datos_3.sort_values(['departamento', 'fecha'])
datos_3
```

```python
datos_3.to_csv('datos_producidos/datos_3.csv', index=False)
```
