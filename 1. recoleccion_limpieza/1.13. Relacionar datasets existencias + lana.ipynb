Observamos el contenido del dataset de existencias 1965-2017

```python
existencias_chubut_1 = pd.read_csv('datos_producidos/existencias_chubut_1965-2017.csv', sep=",")
existencias_chubut_1
```

Eliminamos las columnas que no necesitamos, transformamos la del tiempo en el tipo de dato datetime, seleccionamos la columna de tiempo como indice y luego filtramos para quedarnos con los datos del 2000-2007.

```python
existencias_chubut_1.drop(['provincia', 'hasta_año', 'fuente'], axis = 'columns', inplace=True)
existencias_chubut_1['desde_año'] = pd.to_datetime(existencias_chubut_1['desde_año'], format="%Y").dt.normalize()
existencias_chubut_1 = existencias_chubut_1.set_index(['desde_año'])
existencias_chubut_filtrada=existencias_chubut_1.loc['2000-01-01':'2007-01-01']
existencias_chubut_filtrada
```

Se transformaron los datos de la cantidad de toneladas a kilogramos.

```python
existencias_chubut_filtrada['cantidad'] = existencias_chubut_filtrada['cantidad'].apply(lambda x : x * 1000)
existencias_chubut_filtrada
```

Para poder completar bien la filas faltantes seria interesante saber el porcentaje de corderos, carneros, ovejas, etc, que cada departamento tenia en base a su total de ovinos cada año.

Entonces utilizamos el dataset de existencias de SENASA para obtener los totales anuales.

```python
existencias_anuales = existencias_chubut[existencias_chubut.departamento.eq('TOTAL')]
existencias_anuales
```

```python
totales_chubut = []
for i in existencias_anuales.index:
    totales_chubut.append(
        {'anio': existencias_anuales["anio"][i], 
         'total_ovinos': existencias_anuales["total_ovinos"][i]})

totales_chubut
```

Lo siguiente es recorrer todos los departamentos de chubut y por cada departamento recorrer los totales que obtuvimos antes, luego en base al total de ovinos de departamento (total_ovinos) calculamos el porcentaje de corderos, carneros, etc, y por ultimo en base al total de ovinos de chubut calculamos el porcentaje de ovinos que representa eso para el departamento. De esta manera obtenemos un dataset de porcentajes de Chubut.

```python
departamentos_chubut = ["BIEDMA", "CUSHAMEN", "ESCALANTE", "FLORENTINO AMEGHINO", "FUTALEUFU", "GAIMAN", "GASTRE",
                        "LANGUIÑEO", "MARTIRES", "PASO DE INDIOS", "RAWSON", "RIO SENGUERR",
                        "SARMIENTO", "TEHUELCHES", "TELSEN"]
```

```python
columnas = ['anio', 'provincia', 'departamento', 'carneros', 'ovejas', 'borregos', 'capones',
            'corderos_corderas', 'total_ovinos']
```

```python
def calcular_porcentaje(valor, total):
    return round(valor*100/total, 2)

def calcular_promedio(columna):
    return round(columna.mean)
```

```python
porcentajes_dpto = []

for dpto in departamentos_chubut:
    datos_dpto = existencias_chubut[existencias_chubut.departamento.eq(dpto)]
    for total in totales_chubut:
        datos_dpto_anio = datos_dpto[datos_dpto.anio.eq(total['anio'])]
        total_ovinos = datos_dpto_anio.iloc[0]['total_ovinos']
        datos = datos_dpto_anio.iloc[0]
        porcentaje_carneros = calcular_porcentaje(
            datos['carneros'], total_ovinos)
        porcentaje_ovejas = calcular_porcentaje(datos['ovejas'], total_ovinos)
        porcentaje_borregos = calcular_porcentaje(
            datos['borregos'], total_ovinos)
        porcentaje_capones = calcular_porcentaje(
            datos['capones'], total_ovinos)
        porcentaje_corderos_corderas = calcular_porcentaje(
            datos['corderos_corderas'], total_ovinos)
        porcentaje_total = calcular_porcentaje(
            datos['total_ovinos'], total['total_ovinos'])
        datos_porcentaje = [datos['anio'],
                            datos['provincia'],
                            datos['departamento'],
                            porcentaje_carneros,
                            porcentaje_ovejas,
                            porcentaje_borregos,
                            porcentaje_capones,
                            porcentaje_corderos_corderas,
                            porcentaje_total
                            ]
        porcentajes_dpto.append(datos_porcentaje)

chubut_porcentajes = pd.DataFrame(porcentajes_dpto, columns=columnas)
chubut_porcentajes
```

```python
chubut_porcentajes.to_csv('datos_producidos/chubut_porcentajes.csv', index=False)
```

Para poder generar los valores faltantes de manera aleatoria pero con cierta "estructura" es necesario obtener las medidas estadisticas de los porcentajes anteriores.

```python
def generar_valores_normales(media, desvio):
    valores = []
    for i in range(1,9):
        valores.append(round(random.normalvariate(media_dpto, desvio_dpto),2)) 
    return valores
```

```python
columnas=['departamento', 'dato', 'media', 'desvio_estandar','min', 'max', 'valor_normal_random']
columnas_numericas = ['carneros', 'ovejas', 'borregos', 'capones','corderos_corderas', 'total_ovinos']
promedios_dptos = []

for dpto in departamentos_chubut:
    for columna in columnas_numericas:
        datos_dpto = chubut_porcentajes[chubut_porcentajes.departamento.eq(dpto)]
        media_dpto = round(datos_dpto[columna].mean(),2)
        desvio_dpto = round(datos_dpto[columna].std(),2)
        min_dpto = datos_dpto[columna].min()
        max_dpto = datos_dpto[columna].max()
        valor_normal = generar_valores_normales(media_dpto, desvio_dpto)
        promedios_dptos.append([dpto,columna,media_dpto, desvio_dpto, min_dpto, max_dpto, valor_normal])
chubut_metricas = pd.DataFrame(promedios_dptos, columns=columnas)
chubut_metricas
```

```python
chubut_metricas.to_csv('datos_producidos/chubut_metricas.csv', index=False)
```

Obviamente al usar valores aleatorios la suma de los porcentajes generados no da siempre 100, por lo tanto es necesario hacer un ajuste. Veamos el ejemplo con uno de los datos. Primero obtengamos los datos del total de ovinos de cada departamento.

```python
total_ovinos_dpto = chubut_metricas.loc[chubut_metricas.loc[:, 'dato'] == 'total_ovinos']
total_ovinos_dpto
```

Ahora agrupamos los datos por año

```python
anio_2000 = []
anio_2001 = []
anio_2002 = []
anio_2003 = []
anio_2004 = []
anio_2005 = []
anio_2006 = []
anio_2007 = []
anios = []

for valor_normal in total_ovinos_dpto['valor_normal_random']:
    anio_2000.append(valor_normal[0])
    anio_2001.append(valor_normal[1])
    anio_2002.append(valor_normal[2])
    anio_2003.append(valor_normal[3])
    anio_2004.append(valor_normal[4])
    anio_2005.append(valor_normal[5])
    anio_2006.append(valor_normal[6])
    anio_2007.append(valor_normal[7])
```

Y miramos el ajuste del año 2001

```python
suma = round(sum(anio_2001),2)
print("Antes del ajuste", suma)
resto=0
if suma>100:
    resto = suma - 100
elif suma<100:
    resto = -(100 - suma)
ajuste = round(resto / 15, 4)
anio_2001_ajustado = list(map(lambda x: round(x - ajuste,2), anio_2001))
suma = round(sum(anio_2001_ajustado),0)
print("Despues del ajuste", suma)
```
