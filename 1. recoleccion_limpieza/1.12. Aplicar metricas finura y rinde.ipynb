# 1.12. Aplicar métricas de lana: finura y rinde

```python
import numpy as np
import pandas as pd
import random
```

Después de generar el nuevo archivo con las nuevas métricas para generar los valores, cargamos este archivo csv para convertirlo en un DataFrame y trabajarlo en combinación con el otro más adelante. 

```python
metricas = pd.read_csv("nuevo_porcentajes_f_r.csv")
metricas
```

Lo siguiente es cargar el nuevo Dataset de datos productivos y el clima ajustados con los nuevos valores obtenidos de las estadísticas ganaderas de Chubut, y con una nueva columna de estado para saber cuáles son las filas que deben ser modificadas.

```python
datos = pd.read_csv("actualizando_dataset - Rinde - Finura.csv")
datos
```

Ahora vamos a separar este DataFrame en las filas que son modificables y las que no, obteniendo como resultado un DataFrame de datos modificables y un DataFrame de datos reales, que no tienen que ser modificados. 

```python
df_real = datos[datos['estado'] == 'REAL']
df_real
```

```python
df_modificable = datos[datos['estado'] == 'MODIFICABLE']
df_modificable
```

Creamos una lista con los nombres de las categorías de lana.

```python
categorias_lana = ['finura', 'rinde']
```

Luego creamos una función aleatorio_rango que recibe una media, un desvío, un mínimo y un máximo lo que hacemos es utilizar el random.normalvariate de python y le pasamos una media y un desvío y lo redondeamos a dos; pero mientras ese valor aleatorio que generamos sea mayor al máximo o menor al mínimo vamos a seguir generando otro. 

```python
def aleatorio_rango(media, desvio, minimo, maximo): 
    valor_aleatorio = round(random.normalvariate(media, desvio), 2) 
    while valor_aleatorio < minimo or valor_aleatorio > maximo: 
        valor_aleatorio = round(random.normalvariate(media, desvio), 2) 
    return valor_aleatorio
```

El siguiente paso consiste en recorrer el DataFrame modificable de manera iterativa, extrayendo los departamentos uno por uno y comparándolos con los departamentos presentes en el DataFrame de métricas. A continuación, se recorren las diferentes categorías correspondientes a la finura y al rinde y se obtienen las métricas asociadas, tales como la media, el desvío, el mínimo y el máximo. Luego, se genera un valor aleatorio para cada categoría y se aplica la fórmula correspondiente para obtener el valor calculado. Finalmente, se almacena el valor obtenido en el DataFrame modificable.

```python
for index, row in df_modificable.iterrows():
    departamento = row['departamento']
    df1_departamento = metricas[metricas['departamento'] == departamento] 
    for categoria in categorias_lana:
        media = df1_departamento[categoria].iloc[0]
        desvio = df1_departamento[categoria].iloc[1]
        minimo = df1_departamento[categoria].iloc[2]
        maximo = df1_departamento[categoria].iloc[3]
        valor_aleatorio = aleatorio_rango(media, desvio, minimo, maximo)
        df_modificable.loc[index, categoria] = valor_aleatorio
```

Como resultado obtenemos un nuevo Data Frame con los nuevos valores para la finura y el rinde.

```python
df_modificable
```

Lo siguiente es concatenar los dos dataset: el de datos reales y el de datos modificables para obtener un nuevo Dataset actualizado.

```python
df_actualizado = pd.concat([df_real, df_modificable])
df_actualizado
```

Luego ordenamos el Data frame por el índice. 

```python
df_ordenado = df_actualizado.sort_index()
df_ordenado
```

Y finalmente guardamos el resultado en un archivo csv.

```python
df_ordenado.to_csv("dataset_aplicado_%_lana_f_r.csv", index=False)
```
