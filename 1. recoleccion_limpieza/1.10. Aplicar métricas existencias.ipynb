# 1.10. Aplicar métricas de existencias

```python
import numpy as np
import pandas as pd
import random
```

Después de generar el nuevo archivo con las nuevas métricas para generar los valores, cargamos este archivo csv para convertirlo en un DataFrame y trabajarlo en combinación con el otro más adelante. 

```python
metricas = pd.read_csv("nuevo_porcentajes_existencias.csv")
metricas
```

Lo siguiente es cargar el nuevo Dataset de datos productivos y el clima ajustados con los nuevos valores obtenidos de las estadísticas ganaderas de Chubut, y con una nueva columna de estado para saber cuáles son las filas que deben ser modificadas.

```python
datos = pd.read_csv("actualizando_dataset - Existencias.csv")
datos
```

Ahora vamos a separar este DataFrame en las filas que son modificables y las que no, obteniendo como resultado un DataFrame de datos modificables y un DataFrame de datos reales, que no tienen que ser modificados.

```python
df_real = datos[datos['estado'] == 'REAL']
df_real
```

```python
df_modificable = datos[datos['estado'] == 'MODIFICABLE']
df_modificable
```

Creamos una lista con los nombres de las categorías de ovinos

```python
categorias_ovinos = ['carneros', 'ovejas', 'borregos', 'capones', 'corderos_corderas']
```

Después creamos una función llamada "aleatorio_rango", que toma como entrada una media, una desviación estándar, un valor mínimo y un valor máximo. La función utiliza la función "random.normalvariate" de Python para generar un valor aleatorio con la media y la desviación estándar dadas, y redondea el resultado a dos decimales. Si el valor generado está fuera del rango mínimo y máximo especificado, o si la suma acumulada de los valores generados excede el total de ovinos, la función vuelve a generar otro valor. Esto nos permite ajustar los valores aleatorios generados para que la suma total de los mismos sea exactamente igual al número total de ovinos.

```python
def aleatorio_rango(media, desvio, minimo, maximo, total_ovino, sum_calculado):
    maximo_sum = total_ovino - sum_calculado # El máximo permitido para esta categoría depende de la suma de las categorías anteriores
    valor_aleatorio = round(random.normalvariate(media, desvio), 2)
    while valor_aleatorio < minimo or valor_aleatorio > maximo or sum_calculado + valor_aleatorio > total_ovino:
        valor_aleatorio = round(random.normalvariate(media, desvio), 2)
    return valor_aleatorio
```

El siguiente paso consiste en recorrer el DataFrame modificable de manera iterativa, extrayendo los departamentos uno por uno y comparándolos con los departamentos presentes en el DataFrame de métricas. A continuación, se recorren las diferentes categorías correspondientes a los ovinos y se obtienen las métricas asociadas, tales como la media, el desvío, el mínimo y el máximo. Luego, se genera un valor aleatorio para cada categoría y se aplica la fórmula correspondiente para obtener el valor calculado. Finalmente, se almacena el valor obtenido en el DataFrame modificable.

```python
for index, row in df_modificable.iterrows():
    departamento = row['departamento']
    df1_departamento = metricas[metricas['departamento'] == departamento]
    total_ovino = row['total_ovinos']
    sum_calculado = 0
    
    # Generamos los valores aleatorios para cada categoría de ovinos
    valores = []
    for categoria in categorias_ovinos:
        media = df1_departamento[categoria].iloc[0]
        desvio = df1_departamento[categoria].iloc[1]
        minimo = df1_departamento[categoria].iloc[2]
        maximo = df1_departamento[categoria].iloc[3]
        valor_aleatorio = aleatorio_rango(media, desvio, minimo, maximo, total_ovino, sum_calculado)
        sum_calculado += valor_aleatorio
        valores.append(valor_aleatorio)
    
    # Ajustamos los valores calculados para que la suma sea igual a total_ovino
    factor_ajuste = total_ovino / sum(valores)
    valores_ajustados = [round(valor * factor_ajuste) for valor in valores]
    
    # Ajustamos la última categoría para que la suma total sea exactamente igual al total de ovinos
    diferencia = total_ovino - sum(valores_ajustados)
    valores_ajustados[-1] += diferencia
    
    # Almacenamos los valores calculados en las nuevas columnas del segundo DataFrame
    for i, categoria in enumerate(categorias_ovinos):
        df_modificable.loc[index, categoria] = valores_ajustados[i]
```

Como resultado obtenemos un nuevo DataFrame con los nuevos valores para cada categoría ovina en base al total de ovinos.

```python
df_modificable
```

Lo siguiente es concatenar los dos dataset: el de datos reales y el de datos modificables para obtener un nuevo Dataset actualizado.

```python
df_actualizado = pd.concat([df_real, df_modificable])
df_actualizado
```

Luego ordenamos el Data frame por el índice. 

```python
df_ordenado = df_actualizado.sort_index()
df_ordenado
```

Y finalmente guardamos el resultado en un archivo csv.

```python
df_ordenado.to_csv("dataset_aplicado_%_existencias.csv", index=False)
```
